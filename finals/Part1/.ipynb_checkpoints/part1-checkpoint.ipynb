```python
import time
import datetime 
from datetime import datetime
%matplotlib inline
from matplotlib import pyplot as plt
import csv
import numpy as np
import pandas as pd
import seaborn as sb
```

```python

flights = pd.read_csv('C:/Users/pruth/Desktop/Datasets/flights.csv/flights.csv',low_memory=False)
```

```python
flights.head()
```

```python
flights.shape
```

```python
month_desc={
    1:  '01- January',
    2:  '02- February',
    3:  '03- March',
    4:  '04- April',
    5:  '05- May',
    6:  '06- June',
    7:  '07- July',
    8:  '08- August',
    9:  '09- September',
    10: '10- October',
    11: '11- November',
    12: '12- December'
}
flights['MONTH_NEW'] = flights['MONTH'].apply(lambda m: month_desc[m])


day_of_week_desc = {
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday',
    7: 'Monday'
}
flights['dayofweek'] = flights['DAY_OF_WEEK'].apply(lambda d: day_of_week_desc[d])

```

```python
flights['dayofweek'].head()
```

```python
individualdays = pd.DataFrame(flights.groupby('dayofweek').count()['SCHEDULED_DEPARTURE'])
individualdays = individualdays.sort_values(['SCHEDULED_DEPARTURE'],ascending = 1)
individualdays
```

```python
individualdays.index
```

```python
individualdays.plot(kind='line',figsize=(14,7))
plt.title('Number of flights per day')
plt.ylabel('Number of Departures ------>')
```

```python
individualdays.plot(kind='barh', stacked=True,subplots=True,figsize=(14,7),legend=True);
plt.xlabel('Number of Departures------->')
```

# From this figure we can point out that which day has the maximum number of flights that are scheduled for departure and Friday being the most busy day and Sunday being the least busy.

```python
frequent_month = flights[['MONTH_NEW']].groupby('MONTH_NEW').size().reset_index()
frequent_month.columns = ['MONTH_NEW', 'COUNT']



ax = sb.barplot(x = "MONTH_NEW", y = "COUNT",
           data = frequent_month, palette = "Greens_d")

# ax.light_palette((260, 75, 60), input="husl")


plt.title('Flights per Month')
plt.xlabel('Month------->')
plt.ylabel('Frequency------->')
plt.xticks(rotation='vertical')
plt.show()
```

# The Most number of flights are flown in month of July and least is February
